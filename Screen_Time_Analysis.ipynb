```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("/content/Screentime-App-Details.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

Analyzing the screen time of the user.Looking at the amount of usage of the apps:

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14, 6))

sns.barplot(data=df, x="Date", y="Usage", hue="App")

plt.title("Usage")
plt.xlabel("Date")
plt.ylabel("Usage")
plt.xticks(rotation=45)

plt.show()
```

Looking at the number of notifications from the apps:

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14, 6))

sns.barplot(data=df, x="Date", y="Notifications", hue="App")

plt.title("Notifications")
plt.xlabel("Date")
plt.ylabel("Notifications")
plt.xticks(rotation=45)

plt.show()
```

Looking at the number of times the apps opened:

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14, 6))

sns.barplot(data=df, x="Date", y="Times opened", hue="App")

plt.title("Times Opened")
plt.xlabel("Date")
plt.ylabel("Times Opened")
plt.xticks(rotation=45)

plt.show()
```
